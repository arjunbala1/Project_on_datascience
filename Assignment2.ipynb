# import libraries

```python
#import libraries
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Read data

```python
#reading file
df=pd.read_csv('weather.csv')
df.head(94)
```

```python
#Removing whitespace in Acres and Fires
df['Fires']=df['Fires'].str.replace('Â ','')
df['Acres']=df['Acres'].str.replace('\s+', '', regex=True).astype(str)


```

```python
#Given Acre and Fire values are not Integer,So Converting it into integer for further analysis
df['Acres']=df['Acres'].astype(int)
df['Fires']=df['Fires'].astype(int)
```

# Correlation between Acres and Fires column

```python
#Correlation between Fires and Acres
col1=df['Fires']
col2=df['Acres']
core=col1.corr(col2)
print(core)
```

# Finding Moving Average

```python
#finding moving average
df_fire = pd.DataFrame(df.Fires)
df['MA_fires']=df['Fires'].rolling(5).mean()
df['MA_acres']=df['Acres'].rolling(5).mean()
df.head(94)

```

# relation plot

```python
#Relation plot from 1987 to 2019
start_date=1987
end_date=2019   
mask=(df['Year']>=start_date)&(df['Year']<=end_date)
df=df.loc[mask]
```

```python
plt.figure(figsize=(30,10))
plt.scatter(x=df['Fires'],y=df['Acres'],s=120)
plt.xlabel("Fire",fontsize=20)
plt.ylabel("Acres",fontsize=20)
```
